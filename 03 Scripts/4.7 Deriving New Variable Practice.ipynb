```python
# Define function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df_ords_prods_merge_1['price_range'] = df_ords_prods_merge_1.apply(price_label, axis = 1)
```

## Q2: Defining new columns based on orders_day_of_week

```python
# Creating busy flags for the column based on the orders made of different days of week   
result1 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value in (0,1):
    result1.append("Busiest days")
  elif value in (3,4):
    result1.append("Slowest days")
  else:
    result1.append("Regular days")
```

```python
result1
```

```python
# Assigning the flag to the dataframe as a column
df_ords_prods_merge['busiest_days_modified'] = result1
```

```python
df_ords_prods_merge.head()
```

# Q4: Creating new column based on time of the day

```python
# Counting the value of time_of_day
df_ords_prods_merge['time_of_day'].value_counts()
```

#### We defined most orders if the value counts of the hour is more than 2 million, average orders if between less than 2 million and greater than 4 hundred thousand and fewest orders for less than 4 hundred thousand

```python
# Creating Orders flags for the column based on the orders made on different times of the day   

result2 = []

for value in df_ords_prods_merge["time_of_day"]:
  if value in (10,11,14,15,13,12,16,9,17):
    result2.append("Most orders")
  elif value in (8,18,19,20,7,21,22,23):
    result2.append("Average orders")
  else:
    result2.append("Fewest orders")
```

```python
result2
```

```python
#Assigning the flag as a column to the dataframe
df_ords_prods_merge['busiest_period_of_day'] = result2
```
